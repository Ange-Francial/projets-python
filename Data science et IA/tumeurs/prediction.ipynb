```python
## Séparation des données train et test
```

```python
df = data.copy()
```

```python
df.isna
```

```python
data_train = df.sample(frac=0.7, axis=0)
data_test = df.drop(data_train.index)

x_train = data_train.drop([9], axis=1) # 9 est la colonne cible
y_train = data_train[9]
x_test = data_test.drop([9], axis=1)
y_test = data_test[9]


print('Train : ',data_train.shape,' test : ',data_test.shape)
print('x_Train : ',x_train.shape,' x_Test : ',x_test.shape, 
      '\ny_Train : ',y_train.shape,' y_Test : ',y_test.shape)
```

```python
## Fonction d'activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# création du modèle
n_input = 9 # n_input = x_train.shape[1]
n_hidden = 5 # Nombre de neuronnes de la couche cachée
n_output = 1


model = tf.keras.Sequential([
    tf.keras.layers.Dense(n_hidden, activation = 'sigmoid', input_shape=(n_input,)),
    tf.keras.layers.Dense(n_output, activation = 'sigmoid')
])

#model.summary()
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=1000)
```

```python
score = model.evaluate(x_test, y_test, verbose=1)

#print(f"{model.metrics_names[1]}: {score[1]*100:.2f}")
print('Taux de bonne réponse : {:5.4f}'.format(score[1]))
```
